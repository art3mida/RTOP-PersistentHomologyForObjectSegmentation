```python
import gudhi
import pandas as pd
import numpy as np

%matplotlib inline

from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
from matplotlib.widgets import Slider
```

```python
# Ucitavanje koordinata modela
def read_off(path):
    with open(path, "r") as file:
        if 'OFF' != file.readline().strip():
            raise('Not a valid OFF header')
        n_verts, n_faces, n_dontknow = tuple([int(s) for s in file.readline().strip().split(' ')])
        verts = [[float(s) for s in file.readline().strip().split(' ')] for i_vert in range(n_verts)]
        return verts
    
points = read_off("data/models/eight.off")
points = np.array(points)
```

```python
# Ucitavanje matrice geodezijskih rastojanja
distances = pd.read_csv("data/distance_matricies/eight.csv", header=None, delimiter=',', lineterminator='\n')
distances.shape
```

```python
rips = gudhi.RipsComplex(distance_matrix=distances.values,
                                  max_edge_length=0.15) 
```

```python
st = rips.create_simplex_tree(max_dimension=2)
```

```python
print(st.num_vertices())
print(st.num_simplices())
```

```python
L = st.get_filtration()

triangles = np.array([s[0] for s in L if len(s[0])==3])

fig = plt.figure(figsize=(20,20))
ax = fig.gca(projection='3d')
l = ax.plot_trisurf(points[:,0], points[:,1], points[:,2], triangles=triangles, color="blue")
ax.set_xlim(-0.5, 0.5)
ax.set_ylim(-0.5, 0.5)
ax.set_zlim(-0.5, 0.5)

plt.show()
```
